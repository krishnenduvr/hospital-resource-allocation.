```python
b=a.fillna(method="ffill")
b
```

```python
c=b.fillna(method="bfill")
c
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in c.select_dtypes(include=['object']).columns:
    c[column]=le.fit_transform(c[column])
c
```

```python
c.corr()
```

```python
d=c.drop(["case_id","Hospital_city","patientid","City_Code_Patient"],axis="columns")
d
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
e=d.drop(["Stay_Days"],axis="columns")
e
```

```python
f=d.Stay_Days
f
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(e,f,test_size=0.9)
```

# Logistic Regression

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

# VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
readmission_counts = a["Stay_Days"].value_counts()
labels = readmission_counts.index
plt.figure(figsize=(6, 6))
plt.pie(readmission_counts.values, labels=labels, autopct="%.2f%%", colors=["skyblue", "salmon"])
plt.title("Stay_Days Distribution")
plt.show()
```
